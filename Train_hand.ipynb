```python
import numpy as np 
import pandas as pd 
import os 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data 
df = pd.read_csv("C:/KMUTNB/Project_own/python/Project/hand_sign_classification/dataset_hand.csv",index_col = False)
df.sample(10)
```

```python
# Check missing data
missing_data = df.isnull().sum()
print(missing_data)
df.info()
```

```python
# Prepare data before train model
#import ast
#df["Landmark_Hand"] = df["Landmark_Hand"]
#df["Landmark_Hand"] = df["Landmark_Hand"].apply(lambda x: ast.literal_eval(x))

X = df.drop(["Label"],axis=1)

#X = np.asarray(df.drop(["Label"],axis=1))
y = df["Label"]
X
```

```python
# Split data 

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 47,shuffle=True)

print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
# Train Models
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB


forest_model = RandomForestClassifier()
SVC_model = SVC(kernel="linear",probability=True)
bayes_model = GaussianNB()

forest_model.fit(X_train,y_train)
SVC_model.fit(X_train,y_train)
bayes_model.fit(X_train,y_train)
```

```python
# Evaluate model 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

y_pred_forest = forest_model.predict(X_test)

acc_forest = accuracy_score(y_test,y_pred_forest)

print("Accuracy RD forest:{:.4f}".format(acc_forest))

class_report_forest = classification_report(y_test,y_pred_forest)
CF_matrix_forest = confusion_matrix(y_test,y_pred_forest)
print(class_report_forest)
print(CF_matrix_forest)
```

```python
# Evaluate SVC model

y_pred_svc = SVC_model.predict(X_test)

acc_svc = accuracy_score(y_test,y_pred_svc)
print("Accuracy SVC: {:.4f}".format(acc_svc))

class_report_svc = classification_report(y_test,y_pred_svc)
CF_matrix_svc = confusion_matrix(y_test,y_pred_svc)
print(class_report_svc)
print(CF_matrix_svc)
```

```python
# Evaluate NB model
y_pred_bayes = bayes_model.predict(X_test)

acc_bayes = accuracy_score(y_test,y_pred_bayes)
print("Accuracy NB: {:.4f}".format(acc_bayes))

class_report_bayes = classification_report(y_test,y_pred_bayes)
CF_mattrix_bayes = confusion_matrix(y_test,y_pred_bayes)
print(class_report_bayes)
print(CF_mattrix_bayes)
```
